```python
import pandas as pd

with open("hsa_MTI_strong_positive.csv") as fd:
    miRNA_gene_list = []
    for line in fd:
        line = line.split(",")
        miRNA_name = line[1]  # miRNA name
        gene_name = line[3]   # gene name
        miRNA_gene_list.append([miRNA_name,gene_name])
    
miRNA_gene_list = miRNA_gene_list[1:]   # remove the head 
print(miRNA_gene_list)

print(len(miRNA_gene_list))
# remove the duplicates
miRNA_gene_unique_list = []
for item in miRNA_gene_list:
    if item not in miRNA_gene_unique_list:
        miRNA_gene_unique_list.append(item)

print(len(miRNA_gene_unique_list))

positive_df = pd.DataFrame(miRNA_gene_unique_list,columns=["miRNA_name","gene_name"])

# write to file
# positive_df.to_csv("positive_interaction.csv")




```

```python
import pandas as pd

with open("hsa_tarbase_direct_negative.csv") as fd:
    miRNA_gene_list = []
    for line in fd:
        print (line)
        line = line.split("\t")
        print(line)
        miRNA_name = line[2]  # miRNA name
        gene_name = line[1]   # gene name
        miRNA_gene_list.append([miRNA_name,gene_name])
    
miRNA_gene_list = miRNA_gene_list[1:]   # remove the head 
print(miRNA_gene_list)

print(len(miRNA_gene_list))
# remove the duplicates
miRNA_gene_unique_list = []
for item in miRNA_gene_list:
    if item not in miRNA_gene_unique_list:
        miRNA_gene_unique_list.append(item)

print(len(miRNA_gene_unique_list))

negative_df = pd.DataFrame(miRNA_gene_unique_list,columns=["miRNA_name","gene_name"])

# write to file
negative_df.to_csv("negative_interaction.csv")
```
