```python
import pandas as pd
import os

negative_interaction_df = pd.read_csv("negative_miR_transcript_seq.csv")

# remove the null row
negative_interaction_df = negative_interaction_df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

# negative_interaction_df.to_csv("true_negative_miR_transcript_seq.csv")
print (len(negative_interaction_df))
# print (negative_interaction_df)
i = 0
for index, row in negative_interaction_df.iterrows():
    name = row["shortest_transcript_info"]
    sequence = row["shortest_transcript_seq"]

    fd = open ("temp_seq.txt","w")
    fd.write(name)
    fd.write(sequence)
    fd.close()
        
    predictin_result = os.popen\
         ("cd ../; python3 isMirTarget.py --miRNA %s --mRNA ./test/temp_seq.txt"\
             %(row["miRNA_seq"])).readlines()
    i += 1
    print (i)
    print(row["miRNA_name"])
    print (predictin_result)
    print ("================")
   
    
    
# os.popen("cd ../;python3 isMirTarget.py  --miRNA  %s  --mRNA %s" %())
```
