```python
import os
import pandas as pd

miR_dict = {}
with open("hsa_mature_miR.txt") as fd:
    content_list = fd.readlines()
    for num in range(0,len(content_list)-1,2):
        name_list = content_list[num].split(" ")
        name_alias = name_list[0].replace(">","")
        name_miRD = name_list[-1].strip()
#         print(name_alias)
#         print(name_miRD)
        if name_alias not in miR_dict.keys():
            miR_dict[name_alias] =  name_miRD
# print (miR_dict)


TargetScan7_file_path="/home/zheng/literature_paper/miRNA_target_prediction/Prediction_data_other_softwares/TargetScan7/Predicted_Targets_Info.default_predictions.txt"


def search_TargetScan_result(prediction_result_file,test_data):
    """get the prediction results on test data of other algorithms
    Args: algorithms prediction results, test data
    Return: The number of test data that can be found in the prediction result of a algorithm
    """
    test_df = pd.read_csv(test_data)
    hit_num = 0
    for index, row in test_df.iterrows():
        miRNA_name = row[2]
        miRNA_name_miRD = miR_dict[miRNA_name]
#         print (miRNA_name_miRD)
        transcript_info = row[4]
#         print(transcript_info)
        
        transcript_info_list = transcript_info.split(" ")
#         print (transcript_info_list)
        transcript_ID = transcript_info_list[0].replace(">","")
        transcript_ID = transcript_ID.split(".")[0]       # remove the "." + the number 
        gene_ID = transcript_info_list[3].split(":")[1]
        gene_ID = gene_ID.split(".")[0]
        gene_name = transcript_info_list[6].split(":")[1]
            
#         print (transcript_ID)
#         print (gene_ID)
#         print (gene_name)
#         print("+++")
        result = os.popen("cat %s |grep %s|grep %s" %(prediction_result_file,miRNA_name_miRD,gene_name)).readlines()
        print(result)
        if len(result) > 0:
            hit_num += 1
        
    return hit_num

# hit_num_negative = search_TargetScan_result(TargetScan7_file_path,"true_negative_miR_transcript_seq.csv")
# print (hit_num_negative)

hit_num_postive = search_TargetScan_result(TargetScan7_file_path,"true_positive_miR_transcript_seq.csv")
print (hit_num_postive)

# function search_result(){
#     i=0
#     for line in `cat $1`
#     do  
#         miRNA_name=`echo $line|awk -F ',' '{print $2}'`
#         gene_name=`echo $line|awk -F ',' '{print $3}'`
#         output=`cat $2|grep $miRNA_name|grep $gene_name`
#         echo $output
#         if [ "$output" != "" ]
#         then
#             echo "found"
#             let i=i+1
#         fi
#         echo "+++++++++"    

#     done
#     echo $i
#     return $i
# }

# search_result  "true_negative_miR_gene.csv" $PITA_file_path
# echo $?
```
